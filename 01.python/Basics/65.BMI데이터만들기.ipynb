```python
#BMI 데이터 만들기
#랜덤하게 1명의 이름,키,몸무게
import random
with open("human.txt",'w') as f:
    f.write("이름,키,몸무게\n")
    for i in range(1000):
        한글 = list("가나다라마바사아자차카타파하")
        이름 = random.choice(한글) + random.choice(한글)
        키 = random.randrange(140,200) # v randint is alisas for randrange(a,b+1)
        몸무게 = random.randrange(40,120)
        f.write(f'{이름},{키},{몸무게}\n')

        #print(f'{이름},{키},{몸무게}')  #print(','.join([이름,str(키),str(몸무게)]))
```

```python
#BMI 분석하기
file = open("human.txt","r") # 스트림 연결

for line in file:
    이름,키,몸무게 = line.strip().split(',') #strp은 \n \t등의 공백도 모두 제거함 split()도 마찬가지
    
    if not 키.isdigit():
        continue
    키 = int(키)
    몸무게 = int(몸무게)
    BMI = 몸무게 / (키/100) **2

    if BMI >= 25 : 
        결과 = "과체중"
    elif BMI >= 18.5 :
        결과 = "정상체중"
    else:
        결과 = "저체중"

    결과 
    print(f'이름 : {이름}\n키 : {키}\n몸무게 : {몸무게}\nBMI : {BMI}\n결과 : {결과}')
    print()

file.close()
```
