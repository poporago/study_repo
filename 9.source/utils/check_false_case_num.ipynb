```python
# incorrect_predictions.txt 파일을 등록하여 로깅 내 False case로 기록된 횟수를 count합니다.
num_valid = 20483  #validation데이터셋 샘플 수, 수동 입력
check_file_dir = "C:/Users/ydsc/CustomCRAFT/Weights/saved_models/TPS-ResNet-BiLSTM-Attn-Seed1111_1/incorrect_predictions.txt"
# 검출 횟수를 지정합니다.
num = 50
```

```python
def check_false_case_num(check_file_dir,num):
    dash = '=' * 50
    with open(check_file_dir, 'r') as f :
        lines = f.readlines()
        print(dash)
        text = '전체 False Case의 갯수'
        print(f'{text:^50}')
        print(dash)

        file_dic = {}
        for line in lines:
            iteration, filename, gt, pred, confidence = line.split(',')
            filename = filename.split(' ')[2]

            if filename not in file_dic:
                file_dic[filename] = 1
            file_dic[filename] += 1 

    sorted_num =  sorted(file_dic.items(), key= lambda item:item[1], reverse=True)
    cnt = 0
    for fn in sorted_num:
        if fn[1]> num:
            cnt += 1 
            print(fn)
    print(dash)
    text = '체크 종료, {}회 이상 False Case 샘플의 수는 {}개, 전체 validation갯수 {}개 '.format(num, cnt, num_valid)
    print(f'{text:^50}')
    print(dash)
```

```python
check_false_case_num(check_file_dir,num)
```
