```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# RGB to Grayscale
def rgb_to_grayscale(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Grayscale to RGB
def grayscale_to_rgb(image):
    return cv2.cvtColor(image, cv2.COLOR_GRAY2RGB)

# Example usage
rgb_image = cv2.imread("C:\\Users\\Inbody\\Desktop\\OCR\\data\\My_test\\white_text2.jpg")
grayscale_image = rgb_to_grayscale(rgb_image)
restored_rgb_image = grayscale_to_rgb(grayscale_image)

# Display images
plt.subplot(1, 3, 1)
plt.title('Original RGB')
plt.imshow(cv2.cvtColor(rgb_image, cv2.COLOR_BGR2RGB))

plt.subplot(1, 3, 2)
plt.title('Grayscale')
plt.imshow(grayscale_image, cmap='gray')

plt.subplot(1, 3, 3)
plt.title('Restored RGB')
plt.imshow(cv2.cvtColor(restored_rgb_image, cv2.COLOR_BGR2RGB))

plt.show()
```
