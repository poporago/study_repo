```python
import matplotlib.pyplot as plt
import numpy as np

# 임의의 이미지 데이터 생성 (예: 8x8 픽셀의 MNIST와 유사한 이미지)
image = np.random.randint(0, 256, (8, 8))

# 이미지 시각화
plt.imshow(image, cmap='gray', interpolation='nearest')
plt.title('Image with Pixel Values')
plt.colorbar()  # 색상 막대 추가

# 각 픽셀에 값 표시
for i in range(image.shape[0]):  # 행
    for j in range(image.shape[1]):  # 열
        plt.text(j, i, str(image[i, j]), ha='center', va='center', color='red')

plt.axis('off')  # 축 정보 끄기
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

# 이미지 읽기
image = plt.imread("C:\\Users\\Inbody\\Desktop\\OCR\\발표자료\\logo.png")

# 이미지 크기 얻기
height, width = image.shape[:2]

# 이미지 축소 (가독성을 위해 32배 축소)
scaled_height = int(height / 64)
scaled_width = int(width / 64)

# 최근접 이웃 보간법을 사용하여 이미지 크기 조정
res = cv2.resize(image, (scaled_width, scaled_height), interpolation=cv2.INTER_NEAREST)

# 원본 이미지 표시
plt.imshow(image)

# 각 픽셀에 값 표시
for i in range(res.shape[0]):  # 행
    for j in range(res.shape[1]):  # 열
        # 픽셀 값 추출 (RGB)
        pixel_value = res[i, j]
        # RGB 값을 문자열로 변환
        pixel_text = f"{pixel_value[0]:.0f},{pixel_value[1]:.0f},{pixel_value[2]:.0f}"
        # 텍스트 위치 조정 (축소된 이미지 크기에 맞춰서 원본 이미지의 좌표 계산)
        text_x = j * 64 + 16  # 각 그리드의 중앙에 텍스트 위치
        text_y = i * 64 + 16
        # 텍스트 표시
        plt.text(text_x, text_y, pixel_text, ha='center', va='center', color='black', fontsize=6)

# 그리드 표시
for i in range(0, height, 32):
    plt.axhline(i, color='black', linewidth=0.5)
for j in range(0, width, 32):
    plt.axvline(j, color='black', linewidth=0.5)

# 마지막 가로선과 세로선을 이미지 크기에 맞게 추가
#plt.axhline(height, color='black', linewidth=0.5)
#plt.axvline(width, color='black', linewidth=0.5)

# 축 정보 끄기
plt.axis('off')

# 이미지 보여주기
plt.show()
```
