```python
# Random walk란?
# 매 사건이 독립적이지만 그 위치는 이전 스텝에 영향을 받는 확률과정(Stochastic Process)의 한 종류
```

```python
import numpy as np
import matplotlib.pyplot as plt


def random_walk(n:int):
    random_walking_vector = np.random.rand(n)
    step_vector = []
    
    for random_element in random_walking_vector:
        if random_element >= 0.5:
            random_element = 1
        else :
            random_element = -1
        
        step_vector.append(random_element)
    print(step_vector)
    walks = np.sum(step_vector)
    print(np.abs(walks))
    
    height_vector = []
    height=0
    height_vector.append(height)
    for step in step_vector:
        height += step
        height_vector.append(height)
    
    fig, ax = plt.subplots(1,1,figsize=(15,6))
    ax.plot(height_vector)
    ax.set_xlabel('step')
    ax.set_ylabel('walk')
    ax.set_title("random walk")
    plt.show()
    
    return    
```

```python
random_walk(100)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def random_walk(n: int):
    random_walking_vector = np.random.rand(n)
    step_vector = []

    for random_element in random_walking_vector:
        if random_element >= 0.5:
            step_vector.append(1)
        else:
            step_vector.append(-1)

    walks = np.cumsum(step_vector)
    
    print("Step Vector:", step_vector)
    print("Final position:", walks[-1])

    fig, ax = plt.subplots(1, 1, figsize=(15, 6))
    ax.plot(walks)
    ax.set_xlabel('step')
    ax.set_ylabel('walk')
    ax.set_title("random walk")

    plt.grid()
    plt.show()

random_walk(100)
```
