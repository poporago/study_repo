```python
from sklearn.datasets import load_boston

dataset = load_boston()
print(type(dataset))
print(dataset.keys())
```

```python
raw = dataset['data']
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(raw)
```

```python
df
```

```python
df.columns = dataset['feature_names']
df["target"] =dataset["target"]
df
```

```python
import torch
import torch.nn as nn

from torch.optim.adam import Adam
```

```python
#config
batch_size = 100
lr = 1e-3
epoch = 200
```

```python
#model

model = nn.Sequential(nn.Linear(13,100),
                      nn.ReLU(),
                      nn.Linear(100,1)
                     )

#data
X = df.iloc[:,:13].values  #Series의 형태를 numpy array로 뽑는다.
Y = df['target'].values


#training strategy
criterion = nn.MSELoss()
optim = Adam(model.parameters(),lr=lr)
```

```python
for i in range(epoch):
    for i in range(len(X)//batch_size): #iteration, update횟수는 전체데이터셋에서 배치사이즈로 나눈만큼, drop_last=True
        start = i*batch_size
        end = start+batch_size
        x = torch.FloatTensor(X[start:end])
        y = torch.FloatTensor(Y[start:end])
        
        optim.zero_grad()
        preds = model(x)
        loss = criterion(preds,y)
        loss.backward()
        optim.step()
        
        
    if epoch% 20 == 0 :
        
        print(f'epoch{epoch} loss:{loss.item()}')
        
```
