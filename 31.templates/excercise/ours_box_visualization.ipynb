```python
import os
import json
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.patches as patches

# 이미지와 annotation이 저장된 디렉토리 경로
base_dir = 'C:\\Users\\ydsc\\testset'
image_dir = os.path.join(base_dir, 'Images')
annotation_dir = os.path.join(base_dir, 'labels')

def visualize_image_with_bboxes(base_dir, index=0):
    image_dir = os.path.join(base_dir, 'Images')
    annotation_dir = os.path.join(base_dir, 'labels')
    
    # 이미지와 annotation 파일 리스트를 얻습니다
    image_files = [f for f in os.listdir(image_dir) if f.endswith(('.jpg', '.png'))]
    annotation_files = [f for f in os.listdir(annotation_dir) if f.endswith('.json')]
    
    if index >= len(image_files):
        print("Index out of range")
        return
    
    image_path = os.path.join(image_dir, image_files[index])
    annotation_path = os.path.join(annotation_dir, os.path.splitext(image_files[index])[0] + '.json')
    
    if not os.path.exists(annotation_path):
        print(f"Annotation file not found for image: {image_files[index]}")
        return
    
    # 이미지와 annotation을 로드합니다
    image = Image.open(image_path)
    with open(annotation_path, 'r') as f:
        annotations = json.load(f)
    
    # matplotlib을 사용하여 이미지를 시각화합니다
    fig, ax = plt.subplots(1)
    ax.imshow(image)
    
    # annotation에서 바운딩박스를 추출하고 그립니다
    for word in annotations['words']:
        vertices = word['boundingBox']['vertices']
        polygon = patches.Polygon([(v['x'], v['y']) for v in vertices], linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(polygon)
    
    plt.show()

# 인덱스를 바꿔가며 시각화합니다
for index in range(len(os.listdir(image_dir))):
    visualize_image_with_bboxes(base_dir, index)


#visualize_image_with_bboxes(base_dir, 1)
```
