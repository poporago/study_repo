- 상태 기반(stateful) 명령어 대신 객체 기반(obeject-oriented) 명령어로 가능
- figure / axes / artist / properties

```python
import numpy as np
x = np.random.rand(15) * 10
x.sort()
y = x**2
plt.plot(x, y, marker = "o", ms=10, label="power")
```

```python
fig, axes = plt.subplots(ncols = 2, figsize=(8,4),
                         gridspec_kw={"wspace" : 0.1},
                         constrained_layout=True)

axes[0].scatter(x,y)
axes[1].scatter(x,y)
for ax in axes : 
    ax.set_title("Plot Visualization")
    ax.set_xlabel('xlabel')

fig.suptitle("슈퍼타이틀")
```
