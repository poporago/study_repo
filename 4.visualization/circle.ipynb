```python
import matplotlib.pyplot as plt

# 원형을 그리기 위한 매개변수 설정
circle_outer = plt.Circle((0.5, 0.5), 0.4, color='blue', fill=True, alpha=0.3)
circle_inner = plt.Circle((0.5, 0.5), 0.25, color='green', fill=True, alpha=0.5)

fig, ax = plt.subplots()  # 새로운 그림과 축을 생성합니다.

# 원형 추가
ax.add_artist(circle_outer)
ax.add_artist(circle_inner)

# 텍스트 추가
ax.text(0.5, 0.5, 'Fine-tuning', verticalalignment='center', horizontalalignment='center', color='green', fontsize=12)
ax.text(0.5, 0.8, 'Transfer Learning', verticalalignment='center', horizontalalignment='center', color='blue', fontsize=12)

# 축 설정
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')  # 동일한 스케일로 축을 설정하여 원이 왜곡되지 않도록 합니다.

# 축 제거
ax.axis('off')

# 시각화
plt.title('Relationship between Transfer Learning and Fine-tuning')
plt.show()

```
