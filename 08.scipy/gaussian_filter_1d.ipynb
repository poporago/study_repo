guassian_filter1d의 sigma에 따른 스무딩 효과 실험

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d

# 랜덤 데이터 생성
np.random.seed(0)
x = np.linspace(0, 10, 100)
y = np.sin(x) + np.random.normal(0, 0.5, x.size)

# 가우시안 필터 적용 (다른 sigma 값)
y_sigma_1 = gaussian_filter1d(y, sigma=1)
y_sigma_3 = gaussian_filter1d(y, sigma=3)
y_sigma_5 = gaussian_filter1d(y, sigma=5)

# 그래프 시각화
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Original Data', alpha=0.6)
plt.plot(x, y_sigma_1, label='Sigma=1')
plt.plot(x, y_sigma_3, label='Sigma=3')
plt.plot(x, y_sigma_5, label='Sigma=5')
plt.legend()
plt.title('Effect of Different Sigma Values on Gaussian Smoothing')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
